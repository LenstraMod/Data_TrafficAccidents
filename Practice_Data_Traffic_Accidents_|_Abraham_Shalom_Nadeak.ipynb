# Visualized and gains insight from  all visualized data

# Research and visualize data for crashes that happen per month

```python
crash_counts = data["crash_month"].value_counts().sort_index()
months = crash_counts.index
counts = crash_counts.values
```

```python
plt.figure(figsize=(10,6))
plt.stem(months, counts)
plt.xlabel("Month")
plt.ylabel("Crashes Count")
plt.title("Number of crashes per month")
plt.xticks(months)
plt.grid(True)
plt.show()
```

### Dari visualisasi data di atas kita dapat melihat rata rata kecelakaan yang terjadi per bulannya pada tiap tahunnya. Dari data tersebut dapat disimpulkan bahwa kecelakaan tertinggi terjadi pada bulan September tiap tahunnya dan kecelakaan terendah terjadi pada bulan februari tiap tahunnya
